Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

Нормальное распределение

```python
x = np.linspace(-5,8,100)
mu = [-2,-1,0,1,2]
sigma = [0.6,0.8,1,1.5,2]
```

```python
norm_rv0 = sts.norm(loc=mu[0], scale=sigma[0])
pdf0 = norm_rv0.pdf(x)
norm_rv1 = sts.norm(loc=mu[1], scale=sigma[1])
pdf1 = norm_rv1.pdf(x)
norm_rv2 = sts.norm(loc=mu[2], scale=sigma[2])
pdf2 = norm_rv2.pdf(x)
norm_rv3 = sts.norm(loc=mu[3], scale=sigma[3])
pdf3 = norm_rv3.pdf(x)
norm_rv4 = sts.norm(loc=mu[4], scale=sigma[4])
pdf4 = norm_rv4.pdf(x)
```

```python
fig = plt.figure(figsize=(9, 5))
ax = plt.axes()
ax.plot(x,pdf0,'red', label = f'm={mu[0]} s={sigma[0]}')
ax.plot(x,pdf1,'orange', label = f'm={mu[1]} s={sigma[1]}')
ax.plot(x,pdf2,'yellow', label = f'm={mu[2]} s={sigma[2]}')
ax.plot(x,pdf3,'green', label = f'm={mu[3]} s={sigma[3]}')
ax.plot(x,pdf4,'blue', label = f'm={mu[4]} s={sigma[4]}')
ax.set_xlabel('x')
ax.set_ylabel('F(x)')
ax.legend()
plt.show
```

Экспоненциальное распределение

```python
x = np.linspace(-1,10,100)
lam = [0.1,0.5,0.7,0.9,2]
```

```python
expon_rv0 = sts.expon(lam[0])
pdf0 = expon_rv0.pdf(x)
expon_rv1 = sts.expon(lam[1])
pdf1 = expon_rv1.pdf(x)
expon_rv2 = sts.expon(lam[2])
pdf2 = expon_rv2.pdf(x)
expon_rv3 = sts.expon(lam[3])
pdf3 = expon_rv3.pdf(x)
expon_rv4 = sts.expon(lam[4])
pdf4 = expon_rv4.pdf(x)
```

```python
fig = plt.figure(figsize=(9, 5))
ax = plt.axes()
ax.plot(x,pdf0,'red', label = f'lambda={lam[0]}')
ax.plot(x,pdf1,'orange', label = f'lambda={lam[1]}')
ax.plot(x,pdf2,'yellow', label = f'lambda={lam[2]}')
ax.plot(x,pdf3,'green', label = f'lambda={lam[3]}')
ax.plot(x,pdf4,'blue', label = f'lambda={lam[4]}')
ax.set_xlabel('x')
ax.set_ylabel('F(x)')
ax.legend()
plt.show
```

Распределение Стьюдента

```python
x = np.linspace(-5,5,10000)
n = [5,10,25,50,10000]
```

```python
t_rv0 = sts.t(n[0])
pdf0 = t_rv0.pdf(x)
t_rv1 = sts.t(n[1])
pdf1 = t_rv1.pdf(x)
t_rv2 = sts.t(n[2])
pdf2 = t_rv2.pdf(x)
t_rv3 = sts.t(n[3])
pdf3 = t_rv3.pdf(x)
t_rv4 = sts.t(n[4])
pdf4 = t_rv4.pdf(x)
```

```python
fig = plt.figure(figsize=(15, 8))
ax = plt.axes()
ax.plot(x,pdf0,'red', label = f'n={n[0]}')
ax.plot(x,pdf1,'orange', label = f'n={n[1]}')
ax.plot(x,pdf2,'black', label = f'n={n[2]}')
ax.plot(x,pdf3,'green', label = f'n={n[3]}')
ax.plot(x,pdf4,'blue', label = f'n={n[4]}')
ax.set_xlabel('x')
ax.set_ylabel('F(x)')
ax.legend()
plt.show
```

Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания  в отрезок $[3.5;7]$, т.е. P(3.5 &lt; $\xi$ &lt; 7).

```python
a = 0
b = 5
uniform_rv = sts.uniform(a, b-a)
```

```python
x = np.linspace(-1,8,100)
pdf = uniform_rv.pdf(x)
plt.plot(x, pdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
print(f'C и математическое ожидание равно {pdf.mean()}')
```

```python
print(f'Вероятность попадания в отрезок равна {(5-3.5)*pdf.mean()}')
```

- Задание 3


Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

Используем формулу Хартли.

```python
import math
```

Условие 1

```python
t_r = 54
t_y = 4
t_g = 28
t = t_r + t_y + t_g
t
```

```python
p_r = t_r/t
p_y = t_y/t
p_g = t_g/t
print (f'Вероятность того, что светофор горит красным = {p_r}')
print (f'Вероятность того, что светофор горит желтым = {p_y}')       
print (f'Вероятность того, что светофор горит зеленым = {p_g}')       
```

```python
I = p_r*math.log(2, 1/p_r) + p_y*math.log(2, 1/p_y) + p_g*math.log(2, 1/p_g)
print (f'Количество информации, которое несёт сообщение о цвете светофора в текущий момент = {I}')  
```

Условие 1

```python
t_c = 28
t = 3* t_c
t
```

```python
p_c = t_c/t
print (f'Вероятность того, что светофор горит каким-либо цветом = {p_c}')
```

```python
I = 3 * p_c * math.log(2, 1/p_c)
print (f'Количество информации, которое несёт сообщение о цвете светофора в текущий момент = {I}')  
```
