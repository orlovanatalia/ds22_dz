```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative
from scipy.optimize import fmin
from sympy import *
from pylab import rcParams
rcParams['figure.figsize'] = 15,8
```

### Уровень 0:
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
x = symbols('x')
```

```python
f = x**5 + 4*sin(2*x) + cos(3*x+3)
```

```python
f
```

```python
df1_dx = diff(f,x)
df1_dx
```

```python
f1 = df1_dx.subs(x,1).evalf()
print(f'Первая производная f(x) при x = 1 равна {f1}')
```

```python
df2_dx = diff(df1_dx,x)
df2_dx
```

```python
f2 = df2_dx.subs(x,1).evalf()
print(f'Вторая производная f(x) при x = 1 равна {f2}')
```

### Уровень 1:
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
def f(x):
    return (np.sin(2*x +1))**5
```

```python
f(0)
```

```python
x = np.linspace(-5,5,100)
```

```python
fx = f(x)
```

```python
f1x = [derivative(f, xx, n=1) for xx in x]
```

```python
f2x = [derivative(f, xx, n=2) for xx in x]
```

```python
plt.plot(x, fx, label = 'function')
plt.plot(x, f1x, label = 'first derivative')
plt.plot(x, f2x, label = 'second derivative')
plt.grid()
plt.legend()
plt.show()
```
